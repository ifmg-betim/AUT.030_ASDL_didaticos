# Sinais de tempo discreto
Apresentamos aqui alguns dos principais sinais reais em tempo discreto.

```python
# Importa bibliotecas
import numpy as np, scipy as sp, matplotlib.pyplot as plt
from ipywidgets import interact, interactive, fixed, interact_manual
import ipywidgets as widgets
%matplotlib inline
```

## Sinais senoidais

- $x[n] = sen(\omega\pi n)$


- $x[n] = cos(\omega\pi n)$


- $x[n] = sinc(n)$


- $x[n] = sen(\omega\pi n) + cos(\omega\pi n)$


- $x[n] = sinc(n) + cos(\omega\pi n)$


- $x[n] = sinc(n) + sen(\omega\pi n)$


```python
# Mostra Sinais discretos no domínio do tempo
def sinalDiscreto(sinal, omega=.1, omega2=.1, n = 11):
    n = int(n)
    N = 501; tf = n; t0 = -n;
    n = np.linspace(t0, tf, len(range(t0 ,tf))+1)
    t = np.linspace(t0, tf, N)
    

    if (sinal=='sin' or sinal=='cos' or sinal=='sinc'):
    
        if sinal == 'sin':
            x_t = np.sin(omega*np.pi*t)
            x_n = np.sin(omega*np.pi*n)

        if sinal == 'cos':
            x_t = np.cos(omega*np.pi*t)
            x_n = np.cos(omega*np.pi*n)

        if sinal == 'sinc':
            x_t = np.sinc(omega*t)
            x_n = np.sinc(omega*n)
            
        plt.figure(figsize=(15,3.5)); plt.title(r'$'+sinal+'('+str(omega)+'\pi n)$')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
    
    if sinal == 'sin+cos':
        x_t = np.sin(omega*np.pi*t)+np.cos(omega2*np.pi*t)
        x_n = np.sin(omega*np.pi*n)+np.cos(omega2*np.pi*n)
        
        plt.figure(figsize=(15,3.5)); plt.title('$sin('+str(omega)+'\pi n)+cos('+str(omega2)+'\pi n)$')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        
    if sinal == 'sinc+cos':
        x_t = np.sinc(omega*t)+np.cos(omega2*np.pi*t)
        x_n = np.sinc(omega*n)+np.cos(omega2*np.pi*n)
        
        plt.figure(figsize=(15,3.5)); plt.title('$sinc('+str(omega)+'\pi n)+cos('+str(omega2)+'\pi n)$')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        
    if sinal == 'sinc+sin':
        x_t = np.sinc(omega*t)+np.sin(omega2*np.pi*t)
        x_n = np.sinc(omega*n)+np.sin(omega2*np.pi*n)
        
        plt.figure(figsize=(15,3.5)); plt.title('$sinc('+str(omega)+'\pi n)+sin('+str(omega2)+'\pi n)$')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
    
# Mecanismo de interação
interact(sinalDiscreto, sinal=['sin', 'cos', 'sinc'], omega=(0.00, 10, 0.1), n=(0.00, 100, 1));
interact(sinalDiscreto, sinal=['sin+cos', 'sinc+cos', 'sinc+sin'], omega=(0.00, 10, 0.1), omega2=(0.00, 10, 0.1), n=(0.00, 100, 1));
```

## Exponenciais

- $x[n] = Ce^{an}$


- $x[n] = C \alpha ^{n}$


- $x[n] = Ce^{rt}cos(\omega n)$


```python
# Mostra Sinais discretos no domínio do tempo
def sinalDiscreto2(sinal, a=.1, C=1, n=11):
    n = int(n)
    N = 501; tf = n; t0 = -n;
    n = np.linspace(t0, tf, len(range(t0 ,tf))+1)
    t = np.linspace(t0, tf, N)
    
    if sinal == 'exp':
        x_t = C*np.exp(a*t)
        x_n = C*np.exp(a*n)
        
        plt.figure(figsize=(15,3.5)); plt.title(str(C)+'\u03B5 ^ ('+str(a)+'n)')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        
    if sinal == 'alfa':
        x_t = C*(a**t)
        x_n = C*(a**n)
        
        plt.figure(figsize=(15,3.5)); plt.title(str(C)+'x'+str(a)+'^($n$)')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        
        
def sinalDiscreto3(sinal, r=.1, omega=5, n=11):
    n = int(n)
    N = 501; tf = n; t0 = -n;
    n = np.linspace(t0, tf, len(range(t0 ,tf))+1)
    t = np.linspace(t0, tf, N)
    
    if sinal == 'exp_complex':
        x_t = np.exp(r*t)*np.cos(omega*np.pi*t)
        x_n = np.exp(r*n)*np.cos(omega*np.pi*n)
        
        plt.figure(figsize=(15,3.5)); plt.title('[\u03B5^'+str(r)+'$n] cos($'+str(omega)+'$\pi n)$')
        plt.plot(t, x_t, 'r', lw=1)
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        
    
# Mecanismo de interação
interact(sinalDiscreto2, sinal=['exp', 'alfa'], a=(-2, 2, 0.1), C=(-2, 2, 0.1), n=(0.00, 100, 1));

interact(sinalDiscreto3, sinal=['exp_complex'], r=(-1, 1, 0.1), omega=(0.00, 10, 0.1), n=(0.00, 100, 1));
```

## Impulso
$\delta(n) = \{1, n=0; 0, n\neq0$ 

```python
def sinalDiscreto4(sinal, A=1, d=0, n=11):
    n = int(n)
    tf = n; t0 = -n;
    n = np.linspace(t0, tf, len(range(t0 ,tf))+1)
    x_n = np.linspace(t0, tf, len(range(t0 ,tf))+1)

    #Função Impulso
    for i in range(len(range(t0 ,tf))+1):
        if n[i] == 0 - d:
            x_n[i] = A * 1
        else:
            x_n[i] = 0
    
    if sinal == 'impulso':
    
        plt.figure(figsize=(15,3.5)); plt.title(str(A)+' x $\delta(n+($'+str(d)+'$))$')
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        plt.show()
    
interact(sinalDiscreto4, sinal=['impulso'], A=(0, 10, 1), d=(-10, 10, 1), n=(0.00, 100, 1));
```

## Degrau

$H(n) = \{1, n>=0; 0, n<0$

```python
def sinalDiscreto5(sinal, A=1, d=0, n=11):
    n = int(n)
    N = 501; tf = n; t0 = -n;
    n = np.linspace(t0, tf, len(range(t0 ,tf))+1)
    x_n = np.linspace(t0, tf, len(range(t0 ,tf))+1)

    #Função Degrau
    for i in range(len(range(t0 ,tf))+1):
        if n[i] < 0 - d:
            x_n[i] = 0
        else:
            x_n[i] = A*1
    
    if sinal == 'degrau':
    
        plt.figure(figsize=(15,3.5)); plt.title(str(A)+' x $H(n+($'+str(d)+'$))$')
        plt.stem(n, x_n, 'k', markerfmt='ok', basefmt='k')
        plt.xticks(range(t0-1,tf+1,1))
        plt.xlim((t0, tf)); plt.grid(1, axis='x'); plt.xlabel(r'$n$'); plt.show()
        plt.show()
    
interact(sinalDiscreto5, sinal=['degrau'], A=(0, 10, 1), d=(-10, 10, 1), n=(0.00, 100, 1));
```
